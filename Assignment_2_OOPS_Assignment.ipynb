```python
# 10. How can you have multiple constructors in a Python class? Explain with an example.

'''
Multiple Constructors in Python
Unlike some languages that support method overloading, Python does not allow multiple constructors (multiple __init__ methods) in a single class. However, you can achieve similar functionality using:

Default Parameters
Class Methods as Alternative Constructors (@classmethod)
'''

# 1. Using Default Parameters (Recommended) - can define a single __init__ method with default parameter values to mimic multiple constructors.

class Person:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects using different arguments
person1 = Person()               # Uses default values
person2 = Person("Alice")        # Uses default age
person3 = Person("Bob", 30)      # Custom values

person1.display()  # Output: Name: Unknown, Age: 0
person2.display()  # Output: Name: Alice, Age: 0
person3.display()  # Output: Name: Bob, Age: 30


# 2. Using Class Methods as Alternative Constructors - Python allows defining alternative constructors using @classmethod. This enables multiple ways to create objects.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_birth_year(cls, name, birth_year):
        current_year = 2024
        age = current_year - birth_year
        return cls(name, age)

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects using different constructors
person1 = Person("Alice", 25)            # Standard constructor
person2 = Person.from_birth_year("Bob", 1995)  # Alternative constructor

person1.display()  # Output: Name: Alice, Age: 25
person2.display()  # Output: Name: Bob, Age: 29
```

```python
# 11. What is method overloading, and how is it related to constructors in Python?

''' 
Method Overloading in Python & Its Relation to Constructors
What is Method Overloading?
Method overloading allows multiple methods with the same name but different parameters to exist in a class. In some languages like Java or C++, this is done by defining multiple methods with different parameter lists.

Does Python Support Method Overloading?
No, Python does not support traditional method overloading. If you define multiple methods with the same name, the last one overwrites the previous ones.

Method Overloading in Constructors (__init__)
Since Python does not support true method overloading, you cannot have multiple __init__ methods in a class. Instead, you can simulate overloading using:

Default Parameter Values
Variable-Length Arguments (*args, **kwargs)
'''

# 1. Using Default Parameters (Recommended) : This method provides flexibility by assigning default values to parameters in the constructor.

class Person:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects with different arguments
person1 = Person()               # Uses default values
person2 = Person("Alice")        # Uses default age
person3 = Person("Bob", 30)      # Custom values

person1.display()  # Output: Name: Unknown, Age: 0
person2.display()  # Output: Name: Alice, Age: 0
person3.display()  # Output: Name: Bob, Age: 30


# 2. Using Variable-Length Arguments (*args, **kwargs): another approach is to use *args (for positional arguments) and **kwargs (for keyword arguments) to handle multiple constructor behaviors.

class Person:
    def __init__(self, *args):
        if len(args) == 1:
            self.name = args[0]
            self.age = 0  # Default age
        elif len(args) == 2:
            self.name = args[0]
            self.age = args[1]
        else:
            self.name = "Unknown"
            self.age = 0

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects with different numbers of arguments
person1 = Person()             # No arguments
person2 = Person("Alice")      # One argument
person3 = Person("Bob", 30)    # Two arguments

person1.display()  # Output: Name: Unknown, Age: 0
person2.display()  # Output: Name: Alice, Age: 0
person3.display()  # Output: Name: Bob, Age: 30


```

```python
# 12. Explain the use of the `super()` function in Python constructors. Provide an example.

'''
The super() Function in Python Constructors
The super() function is used in Python to call methods from a parent (superclass) inside a child (subclass). It is especially useful in constructor (__init__) inheritance, allowing a subclass to extend or modify the behavior of the parent class without rewriting all its initialization logic.

Avoids Code Duplication → Reuses parent class attributes and methods.
Maintains Extensibility → Makes code more scalable and maintainable.
Supports Multiple Inheritance → Helps in managing method resolution order (MRO).


Parent Class (Person): Initializes name and age attributes.
Child Class (Employee):
Calls super().__init__(name, age) to initialize name and age (from Person).
Adds salary as an additional attribute.
Method Overriding:
display() in Employee calls super().display() to print name and age from Person, then adds salary.

When extending a parent class while keeping its attributes/methods.
When overriding a method but still needing the parent class's implementation.
When working with multiple inheritance to manage MRO correctly.
'''

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Subclass using super() to call parent constructor
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)  # Calls the constructor of Person
        self.salary = salary  # Additional attribute for Employee

    def display(self):
        super().display()  # Calls Person's display method
        print(f"Salary: {self.salary}")

# Creating an instance of Employee
emp = Employee("Alice", 30, 50000)
emp.display()


```

```python
# 16. How do you prevent a class from having multiple instances by using constructors in Python? Provide an example.

'''
Preventing Multiple Instances of a Class in Python
To ensure that a class has only one instance throughout the program, we can use the Singleton Design Pattern. This can be implemented in Python using constructors in various ways:

Using a Class Attribute
Using __new__ Method
Using a Metaclass

1. Singleton Using a Class Attribute (Recommended)
This method stores the instance in a class variable and returns the same instance for every object creation attempt.

2. Singleton Using a Metaclass
Another way to implement a singleton is by overriding the __call__ method in a metaclass.
'''

# example 1 :

class Singleton:
    _instance = None  # Class attribute to hold the single instance

    def __new__(cls, *args, **kwargs):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

    def __init__(self, value):
        self.value = value

# Creating multiple objects
obj1 = Singleton("First Instance")
obj2 = Singleton("Second Instance")

# Checking if both objects are the same instance
print(obj1 is obj2)  # Output: True (Both refer to the same instance)

# The value of obj2 is also changed because obj1 and obj2 are the same instance
print(obj1.value)  # Output: Second Instance
print(obj2.value)  # Output: Second Instance

# Example 2:

class SingletonMeta(type):
    _instances = {}

    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super().__call__(*args, **kwargs)
        return cls._instances[cls]

class Singleton(metaclass=SingletonMeta):
    def __init__(self, value):
        self.value = value

# Creating multiple objects
obj1 = Singleton("Instance One")
obj2 = Singleton("Instance Two")

print(obj1 is obj2)  # Output: True (Both refer to the same instance)
print(obj1.value)    # Output: Instance Two
print(obj2.value)    # Output: Instance Two

```

```python
# 19. Explain the use of constructor chaining in Python. Provide a practical example.
'''
Constructor Chaining in Python
Constructor chaining refers to the process where a constructor calls another constructor within the same class or from a parent class. This helps in code reuse and ensures that attributes are properly initialized in a hierarchical structure.

Types of Constructor Chaining in Python
Within the Same Class (Calling another constructor using self.__init__())
From a Parent Class (Using super().__init__())

1. Constructor Chaining Within the Same Class
A constructor in a class can call another constructor in the same class to set default values.

2. Constructor Chaining from a Parent Class (super())
If a subclass wants to reuse the constructor from its parent class, it can use super().
'''

# example 1:
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Student {self.name}, Age: {self.age} created.")

    @classmethod
    def from_name(cls, name):
        return cls(name, 18)  # Default age is 18

# Creating instances using different constructors
s1 = Student("Alice", 21)  # Regular constructor
s2 = Student.from_name("Bob")  # Alternative constructor

# Output:
# Student Alice, Age: 21 created.
# Student Bob, Age: 18 created.


# Example 2 :

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person {self.name}, Age: {self.age} created.")

class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)  # Calling parent constructor
        self.grade = grade
        print(f"Student {self.name} is in Grade {self.grade}.")

# Creating an instance
s1 = Student("Charlie", 20, "A")

# Output:
# Person Charlie, Age: 20 created.
# Student Charlie is in Grade A.
```

```python
# 1. What is inheritance in Python? Explain its significance in object-oriented programming.
'''
Inheritance in Python
Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows one class (called the child or subclass) to derive or inherit the properties and behaviors (attributes and methods) of another class (called the parent or superclass). This promotes code reusability, hierarchical classification, and polymorphism.

Significance of Inheritance in OOP
Code Reusability – Instead of rewriting code, a subclass can reuse the functionality of an existing class.
Improved Maintainability – Changes made in the parent class automatically reflect in child classes, reducing redundancy.
Encapsulation & Extensibility – Developers can build on existing classes, adding or modifying functionality as needed.
Polymorphism – Different child classes can override parent class methods, allowing flexibility in behavior.
Logical Class Hierarchies – Encourages a structured, real-world representation of objects.

Types of Inheritance in Python
Single Inheritance – One child class inherits from one parent class.
Multiple Inheritance – A child class inherits from multiple parent classes.
Multilevel Inheritance – A child class inherits from another child class.
Hierarchical Inheritance – Multiple child classes inherit from a single parent class.
Hybrid Inheritance – A combination of two or more types of inheritance.
'''
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        return "Some sound"

# Child class inheriting from Animal
class Dog(Animal):
    def make_sound(self):  # Method Overriding
        return "Bark"

# Creating an instance of Dog
dog = Dog("Buddy")
print(dog.name)        # Output: Buddy
print(dog.make_sound()) # Output: Bark
```

```python
# 2. Differentiate between single inheritance and multiple inheritance in Python. Provide examples for each.
'''
Single Inheritance
Single inheritance occurs when a child class inherits from only one parent class.
This means the child class acquires the attributes and methods of a single base class. It is straightforward and easy to manage.

Multiple Inheritance
Multiple inheritance occurs when a child class inherits from more than one parent class. 
The child class gains properties and methods from multiple sources, making it more flexible but potentially more complex.


Number of Parent Classes:

Single inheritance has one parent class.
Multiple inheritance has two or more parent classes.
Complexity:

Single inheritance is simpler and easier to manage.
Multiple inheritance can be complex if there are conflicts in method resolution.
Code Reusability:

Single inheritance reuses code from one base class.
Multiple inheritance reuses features from multiple classes, making it more powerful.
Method Resolution:

Single inheritance follows a direct parent-child relationship.
Multiple inheritance follows Method Resolution Order (MRO) to resolve conflicts.


'''
# Single inheritance Example. 

class Animal:
    def speak(self):
        return "Animal makes a sound"

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Creating an object of Dog
dog = Dog()
print('Example of Single Inheritance')
print(dog.speak())  # Output: Dog barks




# Multiple Inheritance example.

# Parent Class 1
class Father:
    def show_father(self):
        return "Father's property"

# Parent Class 2
class Mother:
    def show_mother(self):
        return "Mother's property"

# Child class inheriting from both Father and Mother
class Child(Father, Mother):
    def show_child(self):
        return "Child's own property"

# Creating an object of Child
child = Child()

print('Example of Multiple Inheritance')
print(child.show_father())  # Output: Father's property
print(child.show_mother())  # Output: Mother's property
print(child.show_child())   # Output: Child's own property

```

```python
# 4. Explain the concept of method overriding in inheritance. Provide a practical example.

'''
Method Overriding in Inheritance
Method overriding is a feature in object-oriented programming (OOP) where a child class provides a new implementation for a method that is already defined in its parent class. The method in the child class has the same name and same parameters as the method in the parent class, but with a different implementation.

Key Points About Method Overriding:
The child class redefines the method of the parent class.
The method in the child class must have the same name and parameters as in the parent class.
If needed, the overridden method in the parent class can still be accessed using super().
Helps in achieving polymorphism, allowing different behaviors for the same method across different classes.
'''

# Example of Method Overriding in Python

# Parent class
class Animal:
    def speak(self):
        return "Animal makes a sound"

# Child class overriding the method
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Child class overriding the method differently
class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Creating objects
dog = Dog()
cat = Cat()

# Calling the overridden methods
print(dog.speak())  # Output: Dog barks
print(cat.speak())  # Output: Cat meows


# Using super() to Call the Parent Method

class Bird(Animal):
    def speak(self):
        return super().speak() + " but birds chirp"

# Creating object
bird = Bird()
print(bird.speak())  # Output: Animal makes a sound but birds chirp

```

```python
# 9. What is the purpose of the `issubclass()` function in Python? Provide an example.

'''
Purpose of issubclass() Function in Python
The issubclass() function is used to check if a given class is a subclass of another class.
It is particularly useful in inheritance to verify class relationships before performing operations.

class → The class to check.
class_or_tuple → The class (or tuple of classes) to compare against.
Returns True if class is a subclass of class_or_tuple, otherwise returns False.
'''

# Example 1: Checking Single Inheritance

class Animal:
    pass

class Dog(Animal):
    pass

print(issubclass(Dog, Animal))   # True - Dog is a subclass of Animal
print(issubclass(Animal, Dog))   # False - Animal is not a subclass of Dog
print(issubclass(Dog, object))   # True - All classes in Python inherit from object


# Example 2: Checking Against Multiple Classes

class Cat(Animal):
    pass

print(issubclass(Cat, (Animal, Dog)))  # True - Cat is a subclass of Animal
print(issubclass(Dog, (Cat, str)))     # False - Dog is not a subclass of Cat or str



# Example 3: Using issubclass() in a Function

def check_class_relationship(cls):
    if issubclass(cls, Animal):
        print(f"{cls.__name__} is a subclass of Animal")
    else:
        print(f"{cls.__name__} is NOT a subclass of Animal")

check_class_relationship(Dog)  # Output: Dog is a subclass of Animal
check_class_relationship(str)  # Output: str is NOT a subclass of Animal

```

```python
# 10. Discuss the concept of constructor inheritance in Python. How are constructors inherited in child classes?

'''
Constructor Inheritance in Python
In Python inheritance, a child class inherits the constructor (__init__) of its parent class unless it defines its own constructor. However, if the child class defines its own __init__ method, it overrides the parent's constructor.

To retain the functionality of the parent class constructor, the child class must explicitly call it using super().

How Are Constructors Inherited?
If the child class does NOT define __init__, it inherits the parent’s constructor automatically.
If the child class defines its own __init__, it overrides the parent’s constructor.
To use both parent and child class constructors, use super().__init__() inside the child’s constructor.
'''

# Example 1: Constructor Inheritance Without Overriding

class Animal:
    def __init__(self, name):
        self.name = name

# Child class does not define __init__, so it inherits from Animal
class Dog(Animal):
    pass

# Creating an object of Dog
dog = Dog("Buddy")
print(dog.name)  # Output: Buddy


# Example 2: Overriding the Parent Constructor

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):  # Overriding __init__
        self.name = name  # Manually setting name
        self.breed = breed

dog = Dog("Buddy", "Labrador")
print(dog.name)   # Output: Buddy
print(dog.breed)  # Output: Labrador



# Example 3: Using super() to Call the Parent Constructor

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Calling parent constructor
        self.breed = breed

dog = Dog("Buddy", "Labrador")
print(dog.name)   # Output: Buddy (from Animal)
print(dog.breed)  # Output: Labrador (from Dog)



# Example 4: Multiple Inheritance and Constructor Calling

class A:
    def __init__(self):
        print("Constructor of A")

class B:
    def __init__(self):
        print("Constructor of B")

class C(A, B):
    def __init__(self):
        super().__init__()  # Calls the constructor of A (first in MRO)
        print("Constructor of C")
```

```python
# 12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an example using the `abc` module.
'''
Abstract Base Classes (ABCs) in Python
Abstract Base Classes (ABCs) in Python are used to define a common interface for a set of subclasses.
They help enforce that derived classes implement specific methods, ensuring consistency in an inheritance
hierarchy. ABCs are defined using the abc module.

An abstract base class:
Cannot be instantiated directly.
Can define abstract methods that must be implemented by subclasses.
Can also have concrete (non-abstract) methods.

Explanation:
The Shape class is now an abstract base class, inheriting from ABC.
The @abstractmethod decorator ensures that all subclasses implement the area() method.
Attempting to instantiate Shape directly would result in an error.
Circle and Rectangle correctly implement the area() method, so they can be instantiated and used.
'''

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        """Method that must be implemented in subclasses"""
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

# Example usage
# shape = Shape()  # This would raise an error because Shape is abstract

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle Area: {circle.area():.2f}")
print(f"Rectangle Area: {rectangle.area():.2f}")
```

```python
# 18. What is the "diamond problem" in multiple inheritance, and how does Python address it?
'''
The "Diamond Problem" in Multiple Inheritance
The diamond problem occurs in multiple inheritance when a class inherits from two classes that both inherit from a common base class.
This creates an ambiguity about which method or attribute should be inherited from the base class.

Example of the Diamond Problem
Consider the following inheritance structure:
      A
     / \
    B   C
     \ /
      D

Here, D inherits from both B and C, and both B and C inherit from A. If D calls a method from A, 
Python must decide whether to inherit it through B or C.

How Python Solves the Diamond Problem
Python resolves this using Method Resolution Order (MRO) and the C3 Linearization Algorithm.

Method Resolution Order (MRO)
Python follows a depth-first, left-to-right approach while considering inheritance hierarchy.

The super() function and mro() method can help understand the resolution order.

 Interpretation:

D first checks B, since B is listed first in class D(B, C).
If not found, it moves to C.
If still not found, it looks in A.
Finally, it checks the base object class.

'''

# Example of the Diamond Problem
class A:
    def show(self):
        print("A's method")

class B(A):
    def show(self):
        print("B's method")

class C(A):
    def show(self):
        print("C's method")

class D(B, C):  # Multiple inheritance
    pass

obj = D()
obj.show()

# Using super() to Avoid the Diamond Problem

class A:
    def show(self):
        print("A's method")

class B(A):
    def show(self):
        super().show()  # Calls A's method
        print("B's method")

class C(A):
    def show(self):
        super().show()  # Calls A's method
        print("C's method")

class D(B, C):
    def show(self):
        super().show()  # Calls B, then C due to MRO

obj = D()
obj.show()
```

```python
# 19. Discuss the concept of "is-a" and "has-a" relationships in inheritance, and provide examples of each.

'''
1. "Is-A" Relationship (Inheritance)
Definition:
An "is-a" relationship represents inheritance.
A subclass inherits properties and behaviors from a superclass.
It follows the "generalization-specialization" principle (e.g., A Dog is a Animal).


2. "Has-A" Relationship (Composition)
Definition:
A "has-a" relationship represents composition (or aggregation).
Instead of inheriting, one class contains an instance of another class as an attribute.
It follows the "whole-part" principle (e.g., A Car has a Engine).
'''
# Example of "Is-A" Relationship
class Animal:
    def speak(self):
        print("Animals make sounds.")

class Dog(Animal):  # Dog "is a" type of Animal
    def speak(self):
        print("Dog barks.")

dog = Dog()
dog.speak()  # Output: Dog barks

# Example of "Has-A" Relationship

class Engine:
    def start(self):
        print("Engine starts.")

class Car:
    def __init__(self):
        self.engine = Engine()  # Car "has-a" Engine

    def drive(self):
        self.engine.start()
        print("Car is moving.")

car = Car()
car.drive()

```

```python
# 20. Create a Python class hierarchy for a university system. Start with a base class `Person` and create child classes `Student` and `Professor`, each with their own attributes and methods. Provide an example of using these classes in a university context.


# Base Class: Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Child Class: Student
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def study(self):
        print(f"{self.name} is studying {self.major}.")

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}, Major: {self.major}")

# Child Class: Professor
class Professor(Person):
    def __init__(self, name, age, employee_id, department):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.department = department

    def teach(self):
        print(f"Professor {self.name} is teaching in the {self.department} department.")

    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.employee_id}, Department: {self.department}")

# Example Usage
student = Student("Alice", 20, "S12345", "Computer Science")
professor = Professor("Dr. Smith", 45, "P9876", "Mathematics")

# Display information and actions
student.display_info()
student.study()
print()  # Line break

professor.display_info()
professor.teach()
```

```python
# 1. Explain the concept of encapsulation in Python. What is its role in object-oriented programming?
'''
Encapsulation in Python
Encapsulation is one of the core principles of Object-Oriented Programming (OOP). It refers to the practice of
restricting direct access to certain details of an object and only exposing necessary functionalities through methods.

Role of Encapsulation in OOP
Data Hiding → Prevents direct modification of object attributes to ensure data integrity.
Controlled Access → Uses getter and setter methods to control how attributes are accessed and modified.
Improves Code Maintainability → Protects data from unintended changes, making debugging and updates easier.
Encourages Modularity → Keeps related data and behavior together within a class.
'''

# Example of Encapsulation

class Person:
    def __init__(self, name, age):
        self.name = name         # Public attribute
        self._address = "Unknown" # Protected attribute (convention)
        self.__age = age         # Private attribute

    def get_age(self):  # Getter method for private attribute
        return self.__age

    def set_age(self, new_age):  # Setter method for private attribute
        if new_age > 0:
            self.__age = new_age
        else:
            print("Invalid age!")

# Creating an object
person = Person("Alice", 25)

# Accessing public attribute
print(person.name)  # ✅ Output: Alice

# Accessing protected attribute (allowed but not recommended)
print(person._address)  # ✅ Output: Unknown

# Accessing private attribute directly (causes error)
# print(person.__age)  ❌ AttributeError

# Accessing private attribute via getter method
print(person.get_age())  # ✅ Output: 25

# Modifying private attribute using setter method
person.set_age(30)
print(person.get_age())  # ✅ Output: 30



```

```python
# 3. How can you achieve encapsulation in Python classes? Provide an example.

'''
Encapsulation in Python is achieved by controlling access to class attributes and methods using access modifiers 
and getter/setter methods.

Steps to Achieve Encapsulation:
Use Private Attributes → Define attributes with a double underscore (__) to restrict direct access.
Provide Getters and Setters → Use methods (get_* and set_*) to control attribute access.
Use Protected Attributes → Use a single underscore (_) for attributes meant for internal use but not strictly private.
Encapsulate Methods → Define private methods (__method()) for internal logic.

'''

# Example: Encapsulation in a Bank Account System
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number  # Public attribute
        self.__balance = balance  # Private attribute (Encapsulation)

    def get_balance(self):  # Getter method to access private attribute
        return self.__balance

    def deposit(self, amount):  # Setter method to update private attribute
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining balance: ${self.__balance}")
        else:
            print("Insufficient funds or invalid amount!")

# Creating an account
account = BankAccount("12345", 1000)

# Accessing public attribute
print(f"Account Number: {account.account_number}")  # ✅ Accessible

# Accessing private attribute directly (Not allowed)
# print(account.__balance)  ❌ AttributeError: Private attribute

# Correct way: Using getter
print(f"Balance: ${account.get_balance()}")  # ✅ Balance: $1000

# Depositing money
account.deposit(500)  # ✅ Deposited $500. New balance: $1500

# Withdrawing money
account.withdraw(300)  # ✅ Withdrew $300. Remaining balance: $1200




```

```python
# 4. Discuss the difference between public, private, and protected access modifiers in Python.
'''
In Python, access modifiers control the visibility and accessibility of attributes (variables) and 
methods (functions) within a class. They determine whether a class member can be accessed from outside
the class or should be restricted.

1. Public Access Modifier (public)
🔹 Syntax: No special prefix (e.g., self.name)
🔹 Accessibility: Accessible from anywhere (inside and outside the class).
🔹 Usage: Used for attributes and methods that should be open to the user.

2. Protected Access Modifier (protected)
🔹 Syntax: Prefix with a single underscore (_name)
🔹 Accessibility: Can be accessed outside the class, but by convention, it should be treated as internal.
                    It can be accessed in child (subclass) classes.
🔹 Usage: Used when an attribute or method is meant to be used only within the class and subclasses.

3. Private Access Modifier (private)
🔹 Syntax: Prefix with double underscores (__name)
🔹 Accessibility:Cannot be accessed directly outside the class.
                    Can be accessed only within the class.
🔹 Usage: Used when an attribute/method should be completely hidden from outside access.
'''

# 1. Public Access Modifier (public)
class Person:
    def __init__(self, name):
        self.name = name  # Public attribute

    def greet(self):  # Public method
        print(f"Hello, my name is {self.name}.")

# Creating an object
person = Person("Alice")

# Accessing public attributes and methods
print(person.name)  # ✅ Allowed: Alice
person.greet()  # ✅ Allowed: Hello, my name is Alice.



# 2. Protected Access Modifier (protected)


class Person:
    def __init__(self, name, age):
        self.name = name  
        self._age = age  # Protected attribute

    def _display_age(self):  # Protected method
        print(f"{self.name} is {self._age} years old.")

# Creating an object
person = Person("Alice", 25)

# Accessing protected attributes/methods (Allowed but not recommended)
print(person._age)  # ⚠️ Works, but should be treated as private
person._display_age()  # ⚠️ Works, but should be internal




# 3. Private Access Modifier (private)
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number  # Public attribute
        self.__balance = balance  # Private attribute

    def get_balance(self):  # Public method (getter)
        return self.__balance

# Creating an object
account = BankAccount("12345", 1000)

# Accessing public attributes
print(account.account_number)  # ✅ Allowed: 12345

# Accessing private attributes directly (Not allowed)
# print(account.__balance)  ❌ AttributeError: 'BankAccount' object has no attribute '__balance'

# Correct way: Using a public method
print(account.get_balance())  # ✅ Output: 1000


```

```python
# 12. Explain the concept of property decorators in Python and how they relate to encapsulation.

'''
Property decorators (@property) in Python provide a way to define getter, setter, and deleter methods for
encapsulated attributes, allowing controlled access while maintaining a clean interface.

They are commonly used to:
✔ Control attribute access → Prevent direct modification.
✔ Validate data before setting a value → Enforce rules.
✔ Improve code readability → Looks like direct attribute access but uses methods.

🔹 Property Decorators Syntax
Python provides three decorators to manage access:
1️⃣ @property → Defines a getter method.
2️⃣ @attribute_name.setter → Defines a setter method.
3️⃣ @attribute_name.deleter → Defines a deleter method.


🔹 How Property Decorators Relate to Encapsulation
✔ Hides direct access to private attributes (__balance, __account_number).
✔ Provides controlled access → Only valid operations are allowed (e.g., no negative balance).
✔ Makes code cleaner → No need for explicit get_balance() and set_balance() methods.
✔ Read-Only Attributes → @property without a setter prevents modification (e.g., account_number).

'''

class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance  # Private attribute

    @property
    def balance(self):
        """Getter method: Returns the account balance."""
        return self.__balance

    @balance.setter
    def balance(self, amount):
        """Setter method: Allows deposits but prevents invalid amounts."""
        if amount < 0:
            print("Invalid balance! Cannot be negative.")
        else:
            self.__balance = amount

    @property
    def account_number(self):
        """Getter: Provides read-only access to account number."""
        return self.__account_number  # No setter, making it read-only

# Creating an account
account = BankAccount("123456789", 1000)

# Accessing balance using @property (Looks like an attribute, but it's a method)
print(account.balance)  # ✅ 1000

# Updating balance using @balance.setter
account.balance = 1500  # ✅ Allowed
print(account.balance)  # ✅ 1500

# Attempting to set a negative balance
account.balance = -500  # ❌ Invalid balance! Cannot be negative.

# Trying to modify the account number (Not allowed, no setter defined)
# account.account_number = "987654321"  # ❌ AttributeError: can't set attribute
```

```python
# 1. What is polymorphism in Python? Explain how it is related to object-oriented programming.
'''
Polymorphism is a fundamental concept in Object-Oriented Programming (OOP) that allows objects of different classes 
to be treated as objects of a common superclass. It enables the same interface (method name or operator) to be used 
for different types of objects, improving code flexibility and reusability.

Types of Polymorphism in Python
Method Overriding (Runtime Polymorphism)
Method Overloading (Not natively supported in Python)
Operator Overloading (Compile-time Polymorphism)
Duck Typing (Dynamic Polymorphism)

1. Method Overriding (Runtime Polymorphism)
A child class provides a specific implementation of a method already defined in its parent class.
The method signature (name and parameters) remains the same but behaves differently based on the object type.

2. Method Overloading (Not Natively Supported in Python)
Python does not support traditional method overloading (same method name with different parameters).
However, it can be simulated using default arguments or *args.

3. Operator Overloading (Compile-time Polymorphism)
Python allows operators like +, *, and == to be overloaded to work with user-defined objects by
overriding special methods (dunder methods).

4. Duck Typing (Dynamic Polymorphism)
Python follows duck typing, where the type of an object is determined by its behavior rather than its class.
If an object implements a method, it can be used interchangeably with other objects, regardless of its class.

Polymorphism in OOP
Polymorphism is crucial in OOP as it allows code reusability, flexibility, and scalability.
It enables different objects to respond to the same method in their own way.
Combined with inheritance and encapsulation, polymorphism makes code more maintainable.
'''
# 1. Method Overriding (Runtime Polymorphism)

class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Polymorphic behavior
animals = [Dog(), Cat(), Animal()]

for animal in animals:
    print(animal.speak())  # Calls the overridden method based on the object type

# 2. Method Overloading (Not Natively Supported in Python)

class Calculator:
    def add(self, a, b, c=0):
        return a + b + c  # Works for both two and three arguments

calc = Calculator()
print(calc.add(5, 10))       # Output: 15
print(calc.add(5, 10, 20))   # Output: 35

# 3. Operator Overloading (Compile-time Polymorphism)
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2)  # Output: Vector(6, 8)

# 4. Duck Typing (Dynamic Polymorphism)
class Bird:
    def fly(self):
        return "Bird is flying"

class Airplane:
    def fly(self):
        return "Airplane is flying"

class Fish:
    def swim(self):
        return "Fish is swimming"

# Function using duck typing
def make_it_fly(entity):
    print(entity.fly())  # No need to check the object's class

make_it_fly(Bird())      # Output: Bird is flying
make_it_fly(Airplane())  # Output: Airplane is flying

# make_it_fly(Fish())  # This would cause an AttributeError because Fish does not have a fly() method

```

```python
# 2. Describe the difference between compile-time polymorphism and runtime polymorphism in Python.
'''
Polymorphism in Object-Oriented Programming (OOP) allows a method, function, or operator to perform different tasks based on the context. In Python, polymorphism can be classified into Compile-time Polymorphism and Runtime Polymorphism.

Definition
Compile-time Polymorphism: The method or operator is resolved at compile time.
Runtime Polymorphism: The method is resolved at runtime based on the object type.

Achieved Through
Compile-time Polymorphism: Achieved using operator overloading and method overloading (simulated in Python).
Runtime Polymorphism: Achieved using method overriding in inheritance.

Flexibility
Compile-time Polymorphism: Less flexible, as the method selection is fixed before execution.
Runtime Polymorphism: More flexible, as the method selection depends on the actual object type at runtime.

Performance
Compile-time Polymorphism: Faster execution because method resolution happens before runtime.
Runtime Polymorphism: Slightly slower because the method is determined dynamically at runtime.

Example
Compile-time Polymorphism: Implemented using operator overloading and simulated method overloading with default arguments or *args.
Runtime Polymorphism: Implemented using method overriding in inheritance.

1. Compile-time Polymorphism in Python
Python achieves compile-time polymorphism mainly through operator overloading and method overloading (simulated with default arguments or *args).

2. Runtime Polymorphism in Python
Runtime polymorphism occurs when a method in a subclass overrides a method in the parent class, and 
the method that gets executed depends on the object type at runtime.
'''

# 1. Compile-time Polymorphism in Python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):  # Overloading the + operator
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)  # Output: Vector(6, 8)

# 2. Runtime Polymorphism in Python
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Polymorphism with different objects
animals = [Dog(), Cat(), Animal()]

for animal in animals:
    print(animal.speak())  # Calls the overridden method based on the object type
```

```python
# 3. Create a Python class hierarchy for shapes (e.g., circle, square, triangle) and demonstrate polymorphism through a common method, such as `calculate_area()`.

'''
Explanation
Base Class (Shape)
Defines an abstract method calculate_area(), forcing subclasses to implement it.
Derived Classes (Circle, Square, Triangle)
Each subclass provides its own implementation of calculate_area().
Polymorphism in Action
A list of different shape objects (shapes) is created.
The calculate_area() method is called on each object without knowing its specific type.
Python dynamically invokes the correct implementation based on the object type.
'''
import math

# Base class (Parent)
class Shape:
    def calculate_area(self):
        raise NotImplementedError("Subclasses must implement this method")

# Derived class for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

# Derived class for Square
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def calculate_area(self):
        return self.side ** 2

# Derived class for Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

# Polymorphic behavior
shapes = [Circle(5), Square(4), Triangle(6, 3)]

for shape in shapes:
    print(f"Area of {shape.__class__.__name__}: {shape.calculate_area()}")

```

```python
# 4. Explain the concept of method overriding in polymorphism. Provide an example.

'''
Method overriding is a feature of runtime polymorphism in which a subclass provides a specific implementation of
a method that is already defined in its parent class. The method in the child class must have the same name 
and signature as the method in the parent class.
When an overridden method is called using a subclass object, Python executes the child class’s version of the method, not the parent’s.

Key Points of Method Overriding
The child class redefines a method from the parent class with a new implementation.
It enables dynamic method dispatch, meaning the correct method is called based on the object type at runtime.
It helps in achieving polymorphism, allowing a common interface for different classes.
'''

# Parent class
class Animal:
    def speak(self):
        return "Animal makes a sound"

# Child class 1
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Child class 2
class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Creating objects
animal = Animal()
dog = Dog()
cat = Cat()

# Demonstrating method overriding
print(animal.speak())  # Output: Animal makes a sound
print(dog.speak())     # Output: Dog barks (Overridden method)
print(cat.speak())     # Output: Cat meows (Overridden method)
```

```python
# 10. What is the role of the `@abstractmethod` decorator in achieving polymorphism in Python? Provide an example.

'''
The @abstractmethod decorator in Python is used to define abstract methods within an abstract class. 
It ensures that any subclass must implement the abstract method, promoting polymorphism by enforcing a common interface across multiple subclasses.

Role of @abstractmethod in Polymorphism
Enforces Method Implementation: Subclasses must provide their own implementation of the abstract method.
Provides a Common Interface: Different subclasses can be used interchangeably while having their own specific implementations.
Achieves Runtime Polymorphism: Objects of different subclasses can be referenced through a common superclass.
'''

from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass  # Abstract method (must be implemented by subclasses)

# Subclass implementing the abstract method
class Dog(Animal):
    def make_sound(self):
        return "Woof!"

# Another subclass implementing the abstract method
class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Polymorphic behavior
animals = [Dog(), Cat()]  # List of different Animal types

for animal in animals:
    print(animal.make_sound())  # Calls respective subclass method

```

```python
# 11. Create a Python class called `Shape` with a polymorphic method `area()` that calculates the area of different shapes (e.g., circle, rectangle, triangle).

from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass  # Must be implemented by subclasses

# Circle subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2  # πr²

# Rectangle subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height  # width × height

# Triangle subclass
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height  # ½ × base × height

# Polymorphic behavior
shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]

for shape in shapes:
    print(f"Area: {shape.area():.2f}")  # Calls the appropriate subclass method
```

```python
# 14. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card) and demonstrate polymorphism by implementing a common `withdraw()` method.

from abc import ABC, abstractmethod

# Abstract base class for all accounts
class BankAccount(ABC):
    def __init__(self, balance):
        self.balance = balance

    @abstractmethod
    def withdraw(self, amount):
        pass  # Must be implemented by subclasses

# Savings Account
class SavingsAccount(BankAccount):
    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds in Savings Account"
        self.balance -= amount
        return f"Savings withdrawal successful. New balance: ${self.balance:.2f}"

# Checking Account
class CheckingAccount(BankAccount):
    def withdraw(self, amount):
        # Checking accounts allow overdrafts up to $500
        if amount > self.balance + 500:
            return "Overdraft limit exceeded in Checking Account"
        self.balance -= amount
        return f"Checking withdrawal successful. New balance: ${self.balance:.2f}"

# Credit Card Account
class CreditCardAccount(BankAccount):
    def withdraw(self, amount):
        # Credit accounts don't have a "balance" but have a credit limit of $3000
        if amount > 3000:
            return "Credit limit exceeded in Credit Card Account"
        return f"Credit Card transaction approved for ${amount:.2f}"

# Demonstrating Polymorphism
accounts = [
    SavingsAccount(1000), 
    CheckingAccount(500), 
    CreditCardAccount(0)  # Credit account starts at 0 but has a limit
]

withdraw_amounts = [200, 800, 2500]  # Different withdrawal attempts

for account, amount in zip(accounts, withdraw_amounts):
    print(account.withdraw(amount))

```

```python
# 15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.
'''
Operator Overloading and Its Relation to Polymorphism in Python
What is Operator Overloading?
Operator overloading in Python allows customizing the behavior of built-in operators (+, -, *, etc.) 
for user-defined classes. It is a form of polymorphism, where the same operator behaves differently depending on the operand types.

For example, + can mean:
Addition (5 + 3 → 8)
String Concatenation ("Hello" + "World" → "HelloWorld")
Custom behavior in a user-defined class
'''

# Example 1: Overloading + Operator (Addition for a Custom Class)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Using overloaded + operator
p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2  # Calls __add__ method
print(p3)  # Output: (6, 8)


# Example 2: Overloading * Operator (Multiplication for a Custom Class)

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __mul__(self, bonus):
        return self.salary * bonus

    def __str__(self):
        return f"{self.name} earns ${self.salary} per month."

# Using overloaded * operator
emp = Employee("Alice", 5000)
new_salary = emp * 2  # Calls __mul__ method
print(f"New Salary: ${new_salary}")  # Output: New Salary: $10000
```

```python
# 17. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement polymorphism through a common `calculate_salary()` method.

from abc import ABC, abstractmethod

# Abstract base class
class Employee(ABC):
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    @abstractmethod
    def calculate_salary(self):
        pass  # Must be implemented by subclasses

# Manager class (fixed bonus)
class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_salary(self):
        return self.base_salary + self.bonus  # Base salary + fixed bonus

# Developer class (hourly rate * hours worked)
class Developer(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        super().__init__(name, 0)  # No fixed base salary
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked  # Pay based on hours worked

# Designer class (base salary + project commission)
class Designer(Employee):
    def __init__(self, name, base_salary, commission):
        super().__init__(name, base_salary)
        self.commission = commission

    def calculate_salary(self):
        return self.base_salary + self.commission  # Base salary + commission

# Demonstrating Polymorphism
employees = [
    Manager("Alice", 5000, 2000),       # Base salary + fixed bonus
    Developer("Bob", 50, 160),          # Hourly rate * hours worked
    Designer("Charlie", 4000, 1000)     # Base salary + project commission
]

for employee in employees:
    print(f"{employee.name}'s Salary: ${employee.calculate_salary():.2f}")

```

```python
# 1. What is abstraction in Python, and how does it relate to object-oriented programming?
'''
Abstraction is one of the core principles of Object-Oriented Programming (OOP).
In simple terms:
Abstraction means hiding the complex implementation details and showing only the essential features to the user.
It helps reduce complexity and increase efficiency by letting the user interact with an object at a high level without needing to know the inner workings.

In Python, abstraction is achieved using:
Abstract Classes
Abstract Methods

Abstraction allows you to define a blueprint for a group of related classes. It’s closely related to encapsulation, 
but where encapsulation is about restricting access, abstraction is about simplifying exposure.

'''

from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract class
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started with key.")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started with button.")

# Polymorphic behavior using abstraction
def vehicle_test(vehicle: Vehicle):
    vehicle.start_engine()

vehicle_test(Car())         # Output: Car engine started with key.
vehicle_test(Motorcycle())  # Output: Motorcycle engine started with button.


```

```python
# 3. Create a Python class called `Shape` with an abstract method `calculate_area()`. Then, create child classes (e.g., `Circle`, `Rectangle`) that implement the `calculate_area()` method. Provide an example of using these classes.
'''
Abstract Class Example with Shape, Circle, and Rectangle
Step 1: Create an abstract base class Shape with calculate_area() as an abstract method
Step 2: Implement child classes Circle and Rectangle that provide their own calculate_area() logic
Step 3: Use the classes polymorphically
'''

from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

# Circle Class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

# Rectangle Class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

# Example of using the classes
shapes = [
    Circle(5),
    Rectangle(4, 6)
]

for shape in shapes:
    print(f"{type(shape).__name__} Area: {shape.calculate_area():.2f}")

```

```python
# 4. Explain the concept of abstract classes in Python and how they are defined using the `abc` module. Provide an example.
'''
What is an Abstract Class?
An abstract class in Python is a class that cannot be instantiated directly. It's meant to be inherited by other classes and typically 
contains one or more abstract methods — methods that have no implementation in the base class and must be overridden in child classes.

Why use Abstract Classes?
To define a common interface or template for a group of related classes.
To enforce that certain methods must be implemented in all subclasses.
Supports polymorphism and promotes clean, organized code.

How to Create Abstract Classes (with abc module)
Step-by-step:
Import ABC and abstractmethod from the abc module.
Inherit your class from ABC (Abstract Base Class).
Use @abstractmethod decorator to mark methods as abstract.
'''

from abc import ABC, abstractmethod

# Abstract Base Class
class Animal(ABC):

    @abstractmethod
    def make_sound(self):
        pass

# Concrete Class
class Dog(Animal):
    def make_sound(self):
        print("Bark!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Using the classes
def animal_sound(animal: Animal):
    animal.make_sound()

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Bark!
animal_sound(cat)  # Output: Meow!


```

```python
# 6. Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and providing methods to deposit and withdraw funds.

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private variable using name mangling

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdraw amount or insufficient funds.")

    def get_balance(self):
        return self.__balance  # Provides read-only access if needed

# Usage example
account = BankAccount("Alice", 1000)

account.deposit(500)       # Output: Deposited $500. New balance: $1500
account.withdraw(300)      # Output: Withdrew $300. New balance: $1200
print(account.get_balance())  # Output: 1200

# Trying to access private balance directly (will not work)
# print(account.__balance)  # ❌ AttributeError

```

```python
# 7. Discuss the concept of interface classes in Python and their role in achieving abstraction.

'''
Interface Classes in Python
What is an Interface?
An interface defines a contract — a set of methods that a class must implement, but it doesn’t provide any method implementation itself.

Think of it as:
“If a class says it implements this interface, then it must provide those behaviors.”

Does Python Have Interfaces Like Java/C#?
Not directly — Python doesn't have built-in interface types like Java or C#. But we can simulate interfaces using:
Abstract Base Classes (ABCs) from the abc module
@abstractmethod decorator

How Interfaces Achieve Abstraction
An interface class lets you:

Define what methods a class should have (without saying how they should work)
Enforce a common structure across multiple unrelated classes
Promote loose coupling and code flexibility
'''

from abc import ABC, abstractmethod

# This is acting like an interface
class Drawable(ABC):
    
    @abstractmethod
    def draw(self):
        pass

# Different implementations
class Circle(Drawable):
    def draw(self):
        print("Drawing a Circle")

class Square(Drawable):
    def draw(self):
        print("Drawing a Square")

# Polymorphic use
def render(shape: Drawable):
    shape.draw()

render(Circle())   # Output: Drawing a Circle
render(Square())   # Output: Drawing a Square
```

```python
# 10. What is the purpose of abstract methods, and how do they enforce abstraction in Python classes?

'''
 What is an Abstract Method?
An abstract method is a method that is declared but not implemented in a class.
In Python, you define an abstract method using the @abstractmethod decorator inside a class that inherits from ABC (Abstract Base Class).

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

Purpose of Abstract Methods
Define a Common Interface
They force all subclasses to follow a specific structure or contract.

Ensure Method Implementation in Subclasses
Any class that inherits from an abstract base class must implement all its abstract methods — or it will also be abstract and can’t be instantiated.

Promote Consistency
Helpful when multiple classes should have similar behaviors but different implementations.
'''

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started.")

# Usage
v1 = Car()
v1.start_engine()  # Output: Car engine started.

v2 = Motorcycle()
v2.start_engine()  # Output: Motorcycle engine started.
```

```python
# 12. Describe the use of abstract properties in Python and how they can be employed in abstract classes.

'''
Abstract Properties?
An abstract property is a property (like a getter/setter) that is declared but not implemented in the base class.
It’s useful when you want subclasses to provide specific property implementations, just like with abstract methods.

To enforce that certain data attributes must be defined in every subclass.
To maintain a consistent interface across multiple classes.
To ensure that derived classes provide custom logic for accessing or modifying an attribute.
'''

from abc import ABC, abstractmethod

class Employee(ABC):

    @property
    @abstractmethod
    def salary(self):
        pass

    @salary.setter
    @abstractmethod
    def salary(self, value):
        pass

# Concrete subclass
class Manager(Employee):
    def __init__(self, base_salary):
        self._salary = base_salary

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value > 0:
            self._salary = value
        else:
            raise ValueError("Salary must be positive.")

# Usage
m = Manager(5000)
print(m.salary)   # Output: 5000
m.salary = 6000
print(m.salary)   # Output: 6000

```

```python
# 13. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement abstraction by defining a common `get_salary()` method.

from abc import ABC, abstractmethod

# Abstract Base Class
class Employee(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def get_salary(self):
        pass

# Concrete Class 1
class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name)
        self.base_salary = base_salary
        self.bonus = bonus

    def get_salary(self):
        return self.base_salary + self.bonus

# Concrete Class 2
class Developer(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        super().__init__(name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def get_salary(self):
        return self.hourly_rate * self.hours_worked

# Concrete Class 3
class Designer(Employee):
    def __init__(self, name, monthly_salary):
        super().__init__(name)
        self.monthly_salary = monthly_salary

    def get_salary(self):
        return self.monthly_salary

# Example usage
employees = [
    Manager("Alice", base_salary=5000, bonus=1500),
    Developer("Bob", hourly_rate=50, hours_worked=160),
    Designer("Charlie", monthly_salary=4000)
]

for emp in employees:
    print(f"{emp.name}'s salary: ${emp.get_salary()}")
```

```python
# 15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.

'''
Abstract Data Type (ADT)?
An Abstract Data Type (ADT) is a theoretical model of a data structure that defines:
What operations can be performed on the data,
But not how those operations are implemented.
In short, ADTs describe behavior, not implementation.

Examples of ADTs:
List
Stack
Queue
Dictionary (Map)
Set


Role of ADTs in Achieving Abstraction
Hide Complexity - Users interact with a simple interface and don't need to know the inner workings.

Encapsulate Data and Operations - Data is manipulated only via defined methods (like append(), pop()), preventing misuse.

Improve Maintainability - Implementation can change internally without affecting the external code that uses it.

Enable Code Reusability - Common operations (like enqueue, dequeue) are reusable across different programs.
'''

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if not self.is_empty():
            return self._items.pop()
        raise IndexError("Pop from empty stack")

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        return None

    def is_empty(self):
        return len(self._items) == 0

# Usage
s = Stack()
s.push(10)
s.push(20)
print(s.pop())  # Output: 20

```

```python
# 18. Explain how abstraction enhances code reusability and modularity in Python programs.
'''
A. Enhancing Code Reusability

Define Once, Reuse Everywhere - Abstract classes define a common interface (e.g., process(), draw()) that can be inherited by many different classes.
Avoids Code Duplication - Shared logic can be placed in abstract or base classes, while subclasses implement only what's unique.
Works with Different Types - You can write generic functions that work with any class implementing a given interface (polymorphism).
Improves Scalability - When a new class is needed (e.g., a new shape, user role, vehicle), you can extend the abstract base without rewriting existing logic.

B. Enhancing Modularity

Separates Interface from Implementation - Abstraction lets you focus on what a class does, not how it does it — promoting cleaner module boundaries.
Easier Maintenance & Updates - Changes in one module’s internal logic won’t affect others, as long as the interface remains unchanged.
Plug-and-Play Components - Components built on abstract interfaces can be interchanged easily, promoting flexibility (e.g., switching databases or UI themes).
Improved Team Collaboration - Teams can work on different modules at once, agreeing on abstract interfaces in advance.
'''

from abc import ABC, abstractmethod

# Abstract class
class NotificationSender(ABC):
    @abstractmethod
    def send(self, message):
        pass

# Concrete classes
class EmailSender(NotificationSender):
    def send(self, message):
        print(f"Sending Email: {message}")

class SMSSender(NotificationSender):
    def send(self, message):
        print(f"Sending SMS: {message}")

# Reusable function
def notify_all(sender: NotificationSender, message):
    sender.send(message)

# Modular & Reusable
email = EmailSender()
sms = SMSSender()

notify_all(email, "Hello via Email!")
notify_all(sms, "Hello via SMS!")
```

```python
# 19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

# Abstract Base Class
class LibrarySystem(ABC):

    @abstractmethod
    def add_book(self, book_title):
        pass

    @abstractmethod
    def borrow_book(self, book_title):
        pass

# Concrete Implementation: City Library
class CityLibrary(LibrarySystem):
    def __init__(self):
        self.books = []

    def add_book(self, book_title):
        self.books.append(book_title)
        print(f"'{book_title}' added to the City Library.")

    def borrow_book(self, book_title):
        if book_title in self.books:
            self.books.remove(book_title)
            print(f"You borrowed '{book_title}' from the City Library.")
        else:
            print(f"Sorry, '{book_title}' is not available.")

# Concrete Implementation: School Library
class SchoolLibrary(LibrarySystem):
    def __init__(self):
        self.books = {"Math": 3, "Science": 2}

    def add_book(self, book_title):
        self.books[book_title] = self.books.get(book_title, 0) + 1
        print(f"'{book_title}' added to the School Library.")

    def borrow_book(self, book_title):
        if self.books.get(book_title, 0) > 0:
            self.books[book_title] -= 1
            print(f"You borrowed '{book_title}' from the School Library.")
        else:
            print(f"Sorry, '{book_title}' is not available.")

# Usage Example
def manage_library(library: LibrarySystem):
    library.add_book("Python Programming")
    library.borrow_book("Python Programming")

# Test with both libraries
print("🔹 City Library:")
city = CityLibrary()
manage_library(city)

print("\n🔹 School Library:")
school = SchoolLibrary()
manage_library(school)


```

```python
# 20. Describe the concept of method abstraction in Python and how it relates to polymorphism.

'''
What is Method Abstraction?
Method abstraction in Python refers to hiding the internal implementation details of a method and only exposing the essential behavior (what it does).
You define what a method should do but not how it does it — this is especially common when using abstract base classes.
It’s a key principle of object-oriented programming (OOP).

How It's Done in Python:
Using the abc module and @abstractmethod decorator.
The abstract class defines method signatures (name + parameters).
Subclasses must implement these abstract methods.


Relates to Polymorphism:
Polymorphism allows the same method (like make_sound()) to behave differently depending on the object calling it.
Because of method abstraction, different classes (like Dog and Cat) can override the abstract method with their own implementation.
This enables writing flexible, extensible, and reusable code.

'''

from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Using abstracted method
def animal_speak(animal: Animal):
    animal.make_sound()

# Polymorphic behavior
animal_speak(Dog())  # Output: Woof!
animal_speak(Cat())  # Output: Meow!
```

```python
# 1. Explain the concept of composition in Python and how it is used to build complex objects from simpler ones.

'''
Composition in Python?
Composition is an object-oriented programming concept where one class contains an object of another class. It’s a way to build 
complex objects by combining simpler ones.

Think of it as a "has-a" relationship — e.g., a Car has a Engine.

How It Works:
You create smaller, reusable classes.
Then, you combine them inside another class to form a more complex system.

Use of Composition?
Promotes modularity and reusability
Makes code easier to manage and test
Helps avoid deep inheritance chains (which can become messy)
'''
# Simple class: Engine
class Engine:
    def start(self):
        print("Engine started.")

# Another simple class: Battery
class Battery:
    def charge(self):
        print("Battery charging...")

# Complex class that uses composition
class ElectricCar:
    def __init__(self):
        self.engine = Engine()     # Composition: has an Engine
        self.battery = Battery()   # Composition: has a Battery

    def start_car(self):
        self.battery.charge()
        self.engine.start()
        print("Electric car is ready to drive!")

# Using the ElectricCar
my_car = ElectricCar()
my_car.start_car()

```

```python
# 2. Describe the difference between composition and inheritance in object-oriented programming.
'''

Composition vs Inheritance in OOP
1. Relationship Type
Inheritance: Represents an "is-a" relationship.
Example: A Dog is a type of Animal.

Composition: Represents a "has-a" relationship.
Example: A Car has an Engine.

2. Code Reuse
Inheritance: Reuses code by extending a base class.

Composition: Reuses code by including instances of other classes.

3. Flexibility
Inheritance: Less flexible — tightly couples child and parent classes.

Composition: More flexible — you can change parts independently.

4. Coupling
Inheritance: Creates strong coupling between classes.

Composition: Encourages looser coupling and better modularity.

5. Reusability
Inheritance: Reuses base class behavior.

Composition: Reuses functionality from multiple classes (more granular).

6. Override Behavior
Inheritance: Child class can override parent methods.

Composition: Behavior is delegated to the included objects.

7. Design Usage
Inheritance: Best when classes are in a clear hierarchy.

Composition: Best when you want to assemble behaviors dynamically.

8. Maintenance
Inheritance: Changes in the base class can affect all derived classes.

Composition: Easier to maintain — changes in one part don't affect others.

9. Multiple Inheritance
Inheritance: Can lead to complexity and issues (like Diamond Problem).

Composition: Avoids those issues — easier to manage multiple behaviors.
'''
```

```python
# 5. How can you implement composition in Python classes? Provide examples of using composition to create complex objects.

'''
Composition in Python is a design principle where one class contains an object of another class as part of its attributes. 
It represents a "has-a" relationship, as opposed to inheritance, which represents an "is-a" relationship.
Composition allows you to build complex objects by combining simpler ones, promoting code reuse and flexibility.
'''

# Simple classes that will be composed
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        return "Engine started"

class Transmission:
    def __init__(self, type):
        self.type = type

    def shift(self):
        return f"Transmission shifted ({self.type})"

# Composed class
class Car:
    def __init__(self, make, model, engine, transmission):
        self.make = make
        self.model = model
        self.engine = engine            # Composition: Car has an Engine
        self.transmission = transmission  # Composition: Car has a Transmission

    def start_car(self):
        print(f"{self.make} {self.model} is starting...")
        print(self.engine.start())
        print(self.transmission.shift())

# Using the composed classes
engine = Engine(300)
transmission = Transmission("Automatic")
my_car = Car("Toyota", "Supra", engine, transmission)

my_car.start_car()

```

```python
# 6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.

# Song class
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration  # in seconds

    def play(self):
        print(f"🎶 Now playing: '{self.title}' by {self.artist} ({self.duration} secs)")

# Playlist class
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []  # Composition: Playlist has Songs

    def add_song(self, song):
        self.songs.append(song)
        print(f"Added '{song.title}' to playlist '{self.name}'.")

    def play_all(self):
        print(f"\n▶️ Playing playlist: {self.name}")
        for song in self.songs:
            song.play()

# MusicPlayer class
class MusicPlayer:
    def __init__(self):
        self.playlists = {}  # Composition: MusicPlayer has Playlists

    def add_playlist(self, playlist):
        self.playlists[playlist.name] = playlist
        print(f"Playlist '{playlist.name}' added to music player.")

    def play_playlist(self, name):
        if name in self.playlists:
            self.playlists[name].play_all()
        else:
            print(f"No playlist named '{name}' found.")

# --- Example usage ---

# Create songs
song1 = Song("Bohemian Rhapsody", "Queen", 354)
song2 = Song("Imagine", "John Lennon", 183)
song3 = Song("Hey Jude", "The Beatles", 431)

# Create a playlist and add songs
chill_playlist = Playlist("Chill Vibes")
chill_playlist.add_song(song1)
chill_playlist.add_song(song2)
chill_playlist.add_song(song3)

# Create music player and add playlist
player = MusicPlayer()
player.add_playlist(chill_playlist)

# Play a playlist
player.play_playlist("Chill Vibes")
```

```python
# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.

'''
Delegation is a design principle where an object handles a request by passing it to (or "delegating" it to) a contained object.
In composition, since an object "has-a" another object, it can delegate behavior or responsibilities to that inner object, 
rather than doing the work itself.

Benefits of Delegation in System Design
Separation of Concerns
Each class focuses on its own job.
You don’t overload one class with unrelated logic.

Reusability
You can reuse the InkCartridge in other systems (like copiers or scanners) without modifying it.

Flexibility
You can swap out or upgrade components easily (e.g., replace a basic cartridge with a smart one).

Simplifies Complex Systems
You build complex behavior by composing simple objects that work together.

Systems become like LEGO blocks—easy to extend, upgrade, or debug.
'''

class InkCartridge:
    def __init__(self, level):
        self.level = level

    def is_empty(self):
        return self.level <= 0

class Printer:
    def __init__(self, cartridge):
        self.cartridge = cartridge

    def print_page(self):
        if self.cartridge.is_empty():  # Delegation
            print("Can't print: Ink is empty!")
        else:
            print("🖨️ Printing page...")
            self.cartridge.level -= 10
```

```python
# 11. How can you encapsulate and hide the details of composed objects in Python classes to maintain abstraction?

'''
Encapsulation and abstraction are core principles of object-oriented design, especially when using composition.
When a class is composed of other objects, you often want to hide the internal details of those objects so users of 
your class interact with a simplified interface.

1. Use Private Attributes
Prefix attributes with _ or __ to indicate they’re internal and shouldn’t be accessed directly.

2. Expose Only What’s Needed (Abstraction)
Provide high-level methods that delegate tasks internally.

3. Hide Complexity Behind Meaningful Interfaces
Even if a class uses multiple components, expose just enough functionality through the main class.

4. Don’t Expose Composed Objects Directly
Avoid returning internal objects unless necessary.
'''

# 1. Use Private Attributes

class Car:
    def __init__(self, engine, transmission):
        self.__engine = engine          # private
        self.__transmission = transmission  # private

    def start(self):
        return self.__engine.start()

    def shift_gear(self):
        return self.__transmission.shift()


# 2. Expose Only What’s Needed (Abstraction)

# The user doesn’t need to know how the engine starts, just call start()
car = Car(engine, transmission)
car.start()         # ✅ Good
car.__engine.start() # ❌ Bad (and technically not accessible due to name mangling)


# 3. Hide Complexity Behind Meaningful Interfaces

class MusicPlayer:
    def __init__(self, playlist):
        self._playlist = playlist  # protected

    def play_all(self):
        self._playlist.play_all()  # Delegation, but hidden inside a clear method


#4. Don’t Expose Composed Objects Directly

# ❌ BAD: Exposes inner object
def get_engine(self):
    return self.__engine

# ✅ BETTER: Just provide needed functionality
def get_horsepower(self):
    return self.__engine.horsepower

```

```python
# 12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.

# Student class
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def info(self):
        return f"{self.name} (ID: {self.student_id})"

# Instructor class
class Instructor:
    def __init__(self, name, department):
        self.name = name
        self.department = department

    def info(self):
        return f"{self.name} - {self.department} Dept."

# CourseMaterial class
class CourseMaterial:
    def __init__(self, title, material_type):
        self.title = title
        self.material_type = material_type  # e.g., "Lecture", "Assignment"

    def info(self):
        return f"{self.title} ({self.material_type})"

# Course class using composition
class Course:
    def __init__(self, course_code, title, instructor):
        self.course_code = course_code
        self.title = title
        self.instructor = instructor           # has-a Instructor
        self.students = []                     # has-a list of Students
        self.materials = []                    # has-a list of CourseMaterials

    def enroll_student(self, student):
        self.students.append(student)
        print(f"Enrolled {student.name} to {self.course_code}")

    def add_material(self, material):
        self.materials.append(material)
        print(f"Added material: {material.title}")

    def course_info(self):
        print(f"\n📘 Course: {self.title} ({self.course_code})")
        print(f"Instructor: {self.instructor.info()}")
        print("👨‍🎓 Enrolled Students:")
        for s in self.students:
            print(f" - {s.info()}")
        print("📚 Course Materials:")
        for m in self.materials:
            print(f" - {m.info()}")

# --- Example usage ---

# Create instructor
instructor = Instructor("Dr. Alice Bennett", "Computer Science")

# Create course
course = Course("CS101", "Intro to Programming", instructor)

# Enroll students
course.enroll_student(Student("Jane Doe", "S1001"))
course.enroll_student(Student("John Smith", "S1002"))

# Add course materials
course.add_material(CourseMaterial("Week 1 Slides", "Lecture"))
course.add_material(CourseMaterial("Assignment 1", "Assignment"))

# Show full course info
course.course_info()
```

```python
# 13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.

'''
Challenges and Drawbacks of Composition
1. Increased Complexity
Why: As you break down systems into many small components, the number of classes and objects increases.
Example: A Car might end up with separate classes for Engine, Transmission, FuelPump, Wheels, etc.
Impact: Can make the code harder to read or onboard into, especially for small teams or new devs.
Mitigation: Use composition thoughtfully—only break things out into classes when it actually improves clarity or reuse.

2. Tight Coupling Between Objects
Why: When one class depends heavily on the internal behavior or structure of another, changes in one can break the other.
Example: If Car accesses low-level details of Engine (like engine.internal_state[2]), any changes to Engine can break Car.
Mitigation: Respect encapsulation. Interact with components only through their public interfaces (methods and properties).

3. Delegation Overhead
Why: You often have to write many methods that just delegate to internal objects (e.g., car.start() → engine.start()).
Impact: Boilerplate code increases. Can feel repetitive or unnecessary in simpler cases.
Mitigation: Use delegation tools (like __getattr__ or utility wrappers) carefully in Python to reduce repetitive code.

4. Object Lifecycle Management
Why: Managing the creation and ownership of many composed objects can be tricky, especially in large applications.
Example: Who is responsible for creating and destroying a DatabaseConnection used inside a Repository?
Mitigation: Use dependency injection or factory patterns to better manage object lifecycles and ownership.

5. Over-Engineering
Why: It’s tempting to "compose everything" and end up with unnecessary layers of abstraction.
Impact: The code becomes harder to understand than a simpler inheritance-based or even procedural version.
Mitigation: Follow the principle: “You ain’t gonna need it” (YAGNI). Don’t overcomplicate unless you have a reason.
'''
```

```python
# 14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

# Ingredient class
class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def info(self):
        return f"{self.quantity} {self.unit} {self.name}"

# Dish class
class Dish:
    def __init__(self, name):
        self.name = name
        self.ingredients = []

    def add_ingredient(self, ingredient):
        self.ingredients.append(ingredient)

    def show_ingredients(self):
        print(f"🍽️ {self.name} Ingredients:")
        for ing in self.ingredients:
            print(f" - {ing.info()}")

# Menu class
class Menu:
    def __init__(self, name):
        self.name = name
        self.dishes = []

    def add_dish(self, dish):
        self.dishes.append(dish)

    def show_menu(self):
        print(f"\n🧾 {self.name} Menu:")
        for dish in self.dishes:
            print(f" - {dish.name}")

# Restaurant class
class Restaurant:
    def __init__(self, name):
        self.name = name
        self.menus = []

    def add_menu(self, menu):
        self.menus.append(menu)

    def display_all_menus(self):
        print(f"\n🏠 Welcome to {self.name}!")
        for menu in self.menus:
            menu.show_menu()
```

```python
# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
'''
Aggregation is a "has-a" relationship where:
One object contains or uses another.
The contained object can exist independently of the container.
It reflects a looser relationship than standard composition.

Example:
A Department has-a list of Professors.
But professors can exist without a department (e.g., they might move between departments).

Strong Composition
Composition is also a "has-a" relationship, but:
The contained object’s lifecycle is tied to the container.
If the container is destroyed, the contained object usually is too.
It reflects a stronger ownership or dependency.

Example:
A Car has-a Engine.
If the car is destroyed, the engine typically isn’t useful on its own anymore.


Aggregation vs Composition – Key Differences

Relationship Type
Aggregation: Weak "has-a" relationship
Composition: Strong "has-a" relationship

Object Ownership
Aggregation: Contained object is not owned exclusively by the parent
Composition: Contained object is fully owned by the parent

Object Lifecycle
Aggregation: Child object can exist independently of the parent
Composition: Child object cannot exist without the parent

Reusability
Aggregation: Child objects can be shared between multiple parents
Composition: Child objects are unique to the parent

Creation of Objects
Aggregation: Objects are often created externally and passed in
Composition: Objects are often created inside the parent class

Example in Real World
Aggregation: A Library has Books (books exist outside the library too)
Composition: A House has Rooms (rooms belong only to that house)

Deletion Behavior
Aggregation: Deleting the parent does not delete the child
Composition: Deleting the parent also deletes the child

Coupling
Aggregation: Looser coupling
Composition: Tighter coupling
'''
# Aggregation Example

class Professor:
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, name):
        self.name = name
        self.professors = []  # Aggregation

    def add_professor(self, professor):
        self.professors.append(professor)

# Professor can exist independently
prof1 = Professor("Dr. Smith")
cs_dept = Department("Computer Science")
cs_dept.add_professor(prof1)


# Composition Example
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Car:
    def __init__(self, model):
        self.model = model
        self.engine = Engine(200)  # Strong Composition

my_car = Car("Tesla Model S")

```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

# Furniture class
class Furniture:
    def __init__(self, name, type_):
        self.name = name
        self.type = type_

    def info(self):
        return f"{self.type}: {self.name}"

# Appliance class
class Appliance:
    def __init__(self, name, function):
        self.name = name
        self.function = function

    def info(self):
        return f"{self.name} (used for {self.function})"

# Room class
class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, item):
        self.furniture.append(item)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def show_contents(self):
        print(f"\n Room: {self.name}")
        if self.furniture:
            print("Furniture:")
            for item in self.furniture:
                print(f"  - {item.info()}")
        else:
            print("  No furniture.")

        if self.appliances:
            print("Appliances:")
            for appliance in self.appliances:
                print(f"  - {appliance.info()}")
        else:
            print("No appliances.")

# House class
class House:
    def __init__(self, address):
        self.address = address
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def show_house(self):
        print(f"\n House at {self.address} has the following rooms:")
        for room in self.rooms:
            room.show_contents()



# Example :-

# Create furniture and appliances
bed = Furniture("King Bed", "Bed")
wardrobe = Furniture("Wooden Wardrobe", "Wardrobe")
tv = Appliance("LED TV", "entertainment")
ac = Appliance("Air Conditioner", "cooling")

# Create a room and add items
bedroom = Room("Master Bedroom")
bedroom.add_furniture(bed)
bedroom.add_furniture(wardrobe)
bedroom.add_appliance(tv)
bedroom.add_appliance(ac)

# Create another room
kitchen = Room("Kitchen")
fridge = Appliance("Refrigerator", "cooling food")
oven = Appliance("Oven", "baking")
kitchen.add_appliance(fridge)
kitchen.add_appliance(oven)

# Create house and add rooms
my_house = House("123 Dreamland Street")
my_house.add_room(bedroom)
my_house.add_room(kitchen)

# Display house contents
my_house.show_house()

```

```python
# 19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?

# 1. Use Setters or Direct Attribute Access

class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

class Car:
    def __init__(self, model, engine):
        self.model = model
        self.engine = engine  # composed object

    def set_engine(self, new_engine):
        self.engine = new_engine  # dynamically replaced

# Example usage
engine1 = Engine(150)
engine2 = Engine(300)

my_car = Car("Sedan", engine1)
print(my_car.engine.horsepower)  # 150

my_car.set_engine(engine2)
print(my_car.engine.horsepower)  # 300 ✅ dynamically changed


# 2. Duck Typing / Interfaces
class GasEngine:
    def start(self):
        print("Starting gas engine...")

class ElectricEngine:
    def start(self):
        print("Starting electric engine silently...")

class Vehicle:
    def __init__(self, engine):
        self.engine = engine

    def start(self):
        self.engine.start()  # delegate call

# Switch engines at runtime
v = Vehicle(GasEngine())
v.start()  # ➡️ Gas engine

v.engine = ElectricEngine()  # ➡️ Dynamic replacement
v.start()  # ➡️ Electric engine


# 3. Dependency Injection (via constructor or setter)
class Weapon:
    def use(self):
        print("Swinging sword!")

class LaserGun:
    def use(self):
        print("Firing laser!")

class Character:
    def __init__(self, weapon):
        self.weapon = weapon

    def attack(self):
        self.weapon.use()

# Flexibility at runtime
hero = Character(Weapon())
hero.attack()  # Sword

hero.weapon = LaserGun()  # Swap weapon dynamically
hero.attack()  # Laser ✅


# 4. Use Factory Methods or Configs to Build Components
def get_engine_by_type(type_):
    if type_ == "electric":
        return ElectricEngine()
    else:
        return GasEngine()

car = Vehicle(get_engine_by_type("electric"))  # chosen at runtime


```

```python
# 20. Create a Python class for a social media application, using composition to represent users, posts, and comments.

from datetime import datetime

# Comment class
class Comment:
    def __init__(self, user, content):
        self.user = user  # Reference to a User
        self.content = content
        self.timestamp = datetime.now()

    def display(self):
        return f"{self.user.username} commented: {self.content} [{self.timestamp.strftime('%Y-%m-%d %H:%M')}]"

# Post class
class Post:
    def __init__(self, user, content):
        self.user = user  # Reference to a User
        self.content = content
        self.timestamp = datetime.now()
        self.comments = []  # Composition

    def add_comment(self, comment):
        self.comments.append(comment)

    def display(self):
        print(f"\n {self.user.username} posted: {self.content} [{self.timestamp.strftime('%Y-%m-%d %H:%M')}]")
        if self.comments:
            print("Comments:")
            for c in self.comments:
                print(f"  - {c.display()}")
        else:
            print("  No comments yet.")

# User class
class User:
    def __init__(self, username):
        self.username = username
        self.posts = []  # Composition

    def create_post(self, content):
        post = Post(self, content)
        self.posts.append(post)
        return post

    def show_posts(self):
        for post in self.posts:
            post.display()

# SocialMediaApp class
class SocialMediaApp:
    def __init__(self):
        self.users = {}  # username: User

    def register_user(self, username):
        if username not in self.users:
            self.users[username] = User(username)
            print(f"Registered user: {username}")
        else:
            print(f" Username '{username}' already taken.")

    def get_user(self, username):
        return self.users.get(username)

    def show_all_posts(self):
        print("\n All Posts in Social Media App:")
        for user in self.users.values():
            user.show_posts()

# Example :

# Create the app
app = SocialMediaApp()

# Register users
app.register_user("alice")
app.register_user("bob")

# Create posts
alice = app.get_user("alice")
bob = app.get_user("bob")

post1 = alice.create_post("Hello world!")
post2 = bob.create_post("Good morning everyone!")

# Add comments
post1.add_comment(Comment(bob, "Hey Alice!"))
post2.add_comment(Comment(alice, "Good morning Bob"))

# Display everything
app.show_all_posts()

```
